```python
from paperbridge.clients import ZoteroClient
from dotenv import load_dotenv
load_dotenv()
import os

ZOTERO_API_KEY = os.getenv("ZOTERO_API_KEY")
ZOTERO_LIBRARY_ID = os.getenv("ZOTERO_LIBRARY_ID")
```

```python
print(f"Key: {ZOTERO_API_KEY}")
print(f"Library ID: {ZOTERO_LIBRARY_ID}")
```

```python
with ZoteroClient(api_key=ZOTERO_API_KEY, library_id=ZOTERO_LIBRARY_ID) as zot:
    results = zot.get_items(limit=5)
```
